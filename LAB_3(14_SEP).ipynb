```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Years_of_Experience': [2, 5, 8, 10, 12, 15, 18, 20],
    'Salary': [50000, 60000, 75000, 80000, 90000, 100000, 110000, 120000]
}


df = pd.DataFrame(data)
print(df)

X = df[['Age', 'Years_of_Experience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)


new_employee_details = [[30, 5]]
predicted_salary = model.predict(new_employee_details)

plt.scatter(df['Age'], df['Salary'], label='Data Points')
plt.plot(df['Age'], model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.scatter(new_employee_details[0][0], predicted_salary, color='green', marker='X', s=100, label='New Employee Prediction')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.title('Employee Salary Prediction')
plt.grid(True)

plt.show()


print(f'Predicted Salary for the new employee: ${predicted_salary[0]:.2f}')

```
